# Logistic Regression


* 머신러닝에서 classifictation을 하는 방법 중 하나이다.
* 기본 원리는 선을 그어서 분류를 하는 방법으로 선을 긋는다는 점에서 linear regression과 닮은 느낌이다.

***
# 수학적 계산!!!!

### 선으로 분류

* 선을 긋는다 == 일차함수 == wx + b

***
* 하지만 우리가 원하는 것은 regression이 아닌 분류 문제. 간단히 말하면 0(C1)과 1(C2)로 분류하는 것 (이진 분류문제로 반드시 0과 1로 분류하는 것은 아님)


* 그래서 우리는 확률을 사용한다. 

* 어떠한 x가 주어졌을 때 p(확률)가 0.5보다 작으면 C2로 p가 0.5보다 크면 C1이 나오는 방식으로 분류를 해야한다.


* 이러한 분류를 위한 확률의 표현 방법으로 odds ratio가 사용된다.
***
### odds ratio

* odds의 단어적 의미 : 도박에서 얻을 확률과 잃을 확률의 비율을 뜻하는 영단어이다.

* odds는 probability의 또 다른 표현법으로 어떠한 사건이 일어날 확률을 p라고 했을 때 odds = p / (1-p)로 표현된다.


* odds ratio는 승산비라고 불리며 어떠한 확률이 될지 안될지에 대한 표현이 가능해서 우리가 C1과 C2를 분류하기 위해서 적합하다.


* ex) p가 0.5를 넘으면 odds비는 1을 넘고 0.5보다 작으면 odds비는 1보다 작다.
***
### logit

* logit = log + odds = logistic + probit(확률을 재는 단위(??)) = log(p/(1-p))


* 멀쩡하게 구한 odds비에 왜 log를 취하는가?????
  * 우리가 원하는 것은 어떠한 값을 직선을 통해서 분류하는 것
  * 그걸 위해서는 직선과 우리가 구한 odds ratio를 매칭시켜야 하는데 범위가 다르다.
  * wx + b의 직선은 (-∞, ∞)의 범위이고
  * odds ratio는 p가 (0, 1)의 범위라서(확률이니깐) (0, ∞)가 된다.
  * odds에 log를 씌우면 범위가 (-∞, ∞)로 실수 전체이다.
  * logit은 odds와 달리 값이 1보다 큰가를 기준으로 하는 게 아니라 0보다 큰가를 기준으로 한다.
![logit](logit.PNG) 
log(p / (1-p)) 함수의 그래프
![odds](odds.PNG)
p / (1-p) 함수의 그래프
***
### sigmoid
* logit = z라고 했을 때 p로 식을 변형하면 
* p(x) = 1 / (1 + e^z)가 되고 이게 sigmoid 함수이다.
* 여기서 z가 wx + b의 선형함수면 우리가 원하는 식이 나온다.
***
### 합치기

* 결국 logit = wx + b를 하면 된다.
* 우리의 목표는 선을 그어서 분류를 하는 것이기 때문에 선형 함수인 wx+b와 분류를 가능케 하는 odds ratio를 이어서 식을 만든다.


* 그걸 p로 정리하게 되면 1 / (1+exp(-z))가 된다. 여기서 z = wx + b 

# 수학적 계산 2
* 우리가 위의 방법으로 얻은 게 결국 뭐냐? 바로 어떠한 값을 분류할 수 있는 최적의 선을 찾으면 그 선을 기반으로 분류를 할 수 있는 방법을 얻었다.

* 그렇다면 우리는 그 직선을 어떻게 해야 찾을 수 있을까???????

* 손실함수(loss function)를 구해서 w와 b값을 찾아야 하는데 이 cost function을 cross-entropy라고 한다.

### cross-entropy
* linear regression에서 w와 b를 구하기 위해서 오차를 SSE로 구해서 그 오차들의 합이 최소가 되는 값을 찾았었다. (미분을 이용해서)

* logistic regression에서는 바로 미분을 사용하지 않을 것이다.

***

* p(C=1|x) = y = sigmoid(Wx + b) 
  * p(C=1|x) : 입력 x에 대해 출력 값이 1일 확률.
  * 이걸 y로 정의하는데 우리가 만들에낸 식인 sigmoid로 나타낼 수 있음
  
  
  
* p(C=0|x) = 1 - p(C=1|x) = 1 - y
  * 반대로 0이 나올 확률은 1-y 이다.
  
***

* 위의 두 식을 묶어서 하나의 식으로 표현하면 다음과 같다
  * ![수식1.png](수식1.png)
  * t=0 or 1이라서 성립한다.
  
***

* 하나의 입력이 아닌 여러개의 데이터가 입력되었을 때 확률을 구한다.
     *   각 확률들은 독립적이므로 각 입력 데이터의 발생 확률을 곱해서 나타낸다.
     *  이 우도함수의 값이 최대가 되는 경우가 가장 정확한 상황을 의미하고 가장 알맞은 w와 b를 찾는 경우를 의미한다.
  
![수식2.png](수식2.png)

* 마지막 수식을 보면 우리가 구한 우도함수에 -log를 취하는 모습을 볼 수 있다.
* -는 값의 최대값이 아닌 최소값을 구하는 문제로 변환시키고
* log를 이용해서 각 수식을 덧셈의 형태로 만들어주면 나중에 미분을 계산할 때 복잡도가 내려간다.

### cross-entropy 결론

* 이렇게 만들어낸 cost function을 최소화하는 값을 찾으면 된다.
* 최소화되는 값을 찾기 위해서는 미분과 gradient descent방법을 사용한다.
