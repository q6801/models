# SVM

```python
%pylab inline

import numpy as np
import matplotlib.pyplot as plt
import scipy.io
```

```python
#  

matfile = scipy.io.loadmat('ex6data1.mat')

X = matfile['X']
y = matfile['y']

pos = np.array([X[i] for i in range(X.shape[0]) if y[i] == 1])
neg = np.array([X[i] for i in range(X.shape[0]) if y[i] == 0])
```

```python
print(X.shape, y.shape)
print(X)
print(y)
```

```python
#  

def plotData():
    plt.figure(figsize=(10,6))
    plt.plot(pos[:, 0],pos[:, 1],'ro',label='Positive Sample')
    plt.plot(neg[:, 0],neg[:, 1],'bo',label='Negative Sample')
    plt.xlabel('Column 1 Variable')
    plt.ylabel('Column 2 Variable')
    plt.legend()
    plt.grid(True)
    
plotData()
```

```python
def costFunction(theta, X, y):
    # theta: (n+1)*1 mat
    # X: m*(n+1) mat
    # y: m*1 mat
    inner = np.dot(X, theta) # m*1 mat
#     print(X)
#     print(theta)
#     print(inner)
    
    # inner 0 (axis=1) 0 insert 
    inner_1 = np.insert(1-inner, 0, 0, axis=1)
    inner_0 = np.insert(1+inner, 0, 0, axis=1)
    
    # 0  bias
#     inner_1[0, 0] = inner_1[0, 1]
#     inner_0[0, 0] = inner_0[0, 1]
    
    cost_1 = np.array(list(map(max, inner_1))).reshape([51, 1])
    cost_0 = np.array(list(map(max, inner_0))).reshape([51, 1])
    
#     print(cost_1.shape, y.shape)
#     print(np.hstack([cost_1, y]))
# #     print(np.hstack([cost_0, y]))
#     print(inner_0)
    
    cost = y*cost_1 + (1-y)*cost_0
    
    return np.sum(cost)
```

```python
def linearEquation(theta, x):
    # theta: (n+1)*1 mat
    bias = theta[0, 0]
    theta = theta[1:, :] # theta: n*1 mat
    y = -theta[0, 0]/theta[1, 0]*x + bias/theta[0, 0]
    return y
```

```python
def plotBoundary(theta):
    plt.figure(figsize=(10,6))
    plt.plot(pos[:, 0],pos[:, 1],'ro',label='Positive Sample')
    plt.plot(neg[:, 0],neg[:, 1],'bo',label='Negative Sample')
    plt.xlabel('Column 1 Variable')
    plt.ylabel('Column 2 Variable')
    plt.legend()
    plt.grid(True)
    
    x = np.linspace(0, 4, 500)
    y = linearEquation(theta, x)
    plt.plot(x, y)
    plt.show()
```

```python
plotBoundary(np.array([[3], [2], [1]]))
```

```python
def gradientDescent(theta, X, y, alpha=0.001, epochs=3000):
    # theta: (n+1)*1 mat
    # X: m*(n+1) mat
    # y: m*1 mat
    tmptheta = theta
    costHist = []
    
    def dcost_dtheta(inner):
        x = inner[1:4]
        y = inner[4]
        if y==1:
            return -x if 1-inner[0]>0 else np.array([0, 0, 0])
        else:
            return x if 1+inner[0]>0 else np.array([0, 0, 0])
    
    for _ in range(epochs):
        inner = np.dot(X, theta) # m*1 mat
#         if aaa: print(inner.shape) # 51 1
            
        # inner = [inner, X, y] m*4 mat
#         inner = np.hstack([inner, X[:, 1:]])
        inner = np.hstack([inner, X])
#         if aaa: print('ix:', inner.shape) # 51 3    51 4
            
        inner = np.hstack([inner, y])
#         if aaa: print('ixy:', inner.shape) # 51 4   51 5
            
        costDiff = np.array(list(map(dcost_dtheta, inner)))
#         if aaa: print('cd:', costDiff.shape) # 51 2   51 3
#         print(costDiff)
            
        costDSum = np.array([[np.sum(costDiff[:, 0])], 
                             [np.sum(costDiff[:, 1])], 
                             [np.sum(costDiff[:, 2])]])
#         if aaa: print('cs:', costDSum.shape)
        
#         if aaa: print('th:', theta.shape)
        tmptheta = theta - alpha * costDSum
#         print(costDiff[0])
        theta = tmptheta
#         if aaa: print('th:', theta.shape)
        costHist.append(costFunction(theta, X, y))
    
#         if _==1: print(costDiff)
        
    
    return theta, costHist
    
```

```python
#  

matfile = scipy.io.loadmat('ex6data1.mat')

X = matfile['X']
y = matfile['y']

pos = np.array([X[i] for i in range(X.shape[0]) if y[i] == 1])
neg = np.array([X[i] for i in range(X.shape[0]) if y[i] == 0])

# ==================

X = np.insert(X, 0, 1, axis=1)

theta = np.array([[-19.], [-3.5], [-4.]])

print('theta0:', theta)
plotBoundary(theta)

theta, costHist = gradientDescent(theta, X, y)

print('theta1:', theta)
plotBoundary(theta)

plt.figure(figsize=(10,6))
plt.plot(range(len(costHist)), costHist, 'bo')
```

```python
def costFunction(theta, X, y):
    # theta: (n+1)*1 mat
    # X: m*(n+1) mat
    # y: m*1 mat
    inner = np.dot(X, theta) # m*1 mat
#     print(X)
#     print(theta)
#     print(inner)
    
    # inner 0 (axis=1) 0 insert 
    inner_1 = np.insert(1-inner, 0, 0, axis=1)
    inner_0 = np.insert(1+inner, 0, 0, axis=1)
    
    # 0  bias
#     inner_1[0, 0] = inner_1[0, 1]
#     inner_0[0, 0] = inner_0[0, 1]
    
    cost_1 = np.array(list(map(max, inner_1))).reshape([51, 1])
    cost_0 = np.array(list(map(max, inner_0))).reshape([51, 1])
    
#     print(cost_1.shape, y.shape)
#     print(np.hstack([cost_1, y]))
# #     print(np.hstack([cost_0, y]))
#     print(inner_0)
    
    cost = y*cost_1 + (1-y)*cost_0
    
    return np.sum(cost)
```
